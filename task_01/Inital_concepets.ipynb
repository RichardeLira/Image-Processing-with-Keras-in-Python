>Import the image from the file bricks.png into data.


>Display the image in data on the screen.


```python
# Import matplotlib
import matplotlib.pyplot as plt
```

```python
# Load the image
data = plt.imread('bricks.jpg')
plt.imshow(data)
plt.show()
```
